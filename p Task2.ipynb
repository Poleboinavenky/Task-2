```python
import pandas as pd

# Load CSV file into a Pandas DataFrame
df = pd.read_csv(r"D:\dt.csv")

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())

# Filtering data based on conditions
filtered_df = df[df['Speed'] > 50]  # Filtering data where speed is greater than 50
print("\nFiltered DataFrame based on condition (Speed > 50):")
print(filtered_df)

# Handling missing values
# Replace missing values in a specific column with a default value
# For example, if you want to replace missing values in the 'Random Event Occurred' column with False
df['Random Event Occurred'].fillna(False, inplace=True)

# Drop rows with missing values
df.dropna(inplace=True)

# Calculating summary statistics
summary_stats = df.describe()
print("\nSummary statistics of the DataFrame:")
print(summary_stats)
```

```python
import pandas as pd

# Load CSV file into a Pandas DataFrame
df = pd.read_csv(r"D:\dt.csv")

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())

# Filtering data based on conditions
filtered_df = df[df['Speed'] > 50]  # Filtering data where speed is greater than 50
print("\nFiltered DataFrame based on condition (Speed > 50):")
print(filtered_df)

# Handling missing values
# Replace missing values in a specific column with a default value
# For example, if you want to replace missing values in the 'Random Event Occurred' column with False
df['Random Event Occurred'] = df['Random Event Occurred'].fillna(False)

# Drop rows with missing values
df.dropna(inplace=True)

# Calculating summary statistics
summary_stats = df.describe()
print("\nSummary statistics of the DataFrame:")
print(summary_stats)
```
